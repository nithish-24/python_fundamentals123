```python
# prompt: create a histogram on sat score column in  the data frame df1

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df1 is already loaded as a pandas DataFrame
# If not, load it first:
# df1 = pd.read_csv('/content/Universities.csv')

plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.hist(df1['SAT'], bins=20, edgecolor='black')  # Increased bins for more detail, added edgecolor
plt.xlabel('SAT Scores', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of SAT Scores', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', alpha=0.5) # Added a grid for better readability
plt.show()
```

```python
# prompt: create a histogram on accept score column in  the data frame df1

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df1 is already loaded as a pandas DataFrame
# If not, load it first:
# df1 = pd.read_csv('/content/Universities.csv')

plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.hist(df1['Accept'], bins=20, edgecolor='black')  # Increased bins for more detail, added edgecolor
plt.xlabel('Acceptance Rate', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Acceptance Rate', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', alpha=0.5) # Added a grid for better readability
plt.show()
```

```python
# prompt: create a histogram on accept column in dataframe df1 using seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df1 is already loaded as a pandas DataFrame
# If not, load it first:
# df1 = pd.read_csv('/content/Universities.csv')

sns.histplot(df1['Accept'])
plt.xlabel('Acceptance Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Acceptance Rate')
plt.show()
```
